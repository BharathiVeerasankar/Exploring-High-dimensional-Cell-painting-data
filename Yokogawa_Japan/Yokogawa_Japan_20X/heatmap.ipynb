# Generating a Heatmap
## 1. Heatmap for Cell number per well

```python
from random import seed
from random import random 
from random import randrange
from matplotlib import pyplot
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
from numpy import linalg as la 
from itertools import chain
import collections
```

```python
df_well=pd.read_csv("normalized.csv",usecols=["Metadata_Well"])
```

```python
df_count=pd.read_csv("normalized.csv",usecols=["Metadata_Count_Cells"])
```

```python
count = df_count.to_numpy()
```

```python
arr_2d = np.reshape(count, (16, 24))
```

```python
# columns = range(24)
# index = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P']
pd.DataFrame(arr_2d).to_csv('count.csv')  
```

```python
count = pd.read_csv('count.csv')
```

```python
count.columns
```

Adapted from : https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html

```python
# figure
fig, ax = plt.subplots(figsize=(8, 6))
# plot heatmap
sns.heatmap(count.drop(['Unnamed: 0'],axis=1), cmap=plt.cm.Oranges)
# xticks
ax.xaxis.tick_top()
yticks_labels = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P']
xticks_labels = range(25)
plt.yticks(np.arange(16) + .5, labels=yticks_labels)
plt.xticks(np.arange(25) + .5, labels=xticks_labels)
# axis labels
plt.xlabel('')
plt.ylabel('')
# title
title = 'Cell Count per well\n'.upper()
plt.title(title, loc='left')
plt.show()
```

It is noted that all the same heatmap is obtained when the other .csv files are used. I.e. all .csv files downloaded have the same well plate and the data obtained from all files are same.
