# Zscore Normalization

### zscore normalized value = (x – μ) / σ

### where: 
 ### x: Original value
 ### μ: Mean of data
 ### σ: Standard deviation of data

```python
area_zscore = stats.zscore(df_area, ddof = 1)
```

```python
area_zscore
```

```python
well = range(384)
plt.plot(well, df_area_normalized, label = 'Normalization technique used in data')  
plt.plot(well, area_zscore, label = 'Zcore Normalization')  
plt.xlabel('Well Number ')
plt.ylabel('Area of cell ')
plt.legend()
plt.show()
```

```python
plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.plot(well, df_area_normalized)
plt.title("Normalization used in data")
plt.xlabel('Well Number ')
plt.ylabel('Area of cell ')

plt.subplot(1, 2, 2) # index 2
plt.plot(well, area_zscore)  
plt.title("Zscore Normalization")
plt.xlabel('Well Number ')
plt.ylabel('Area of cell ')

plt.show()
```

```python
a = np.arange(-2,6)
plt.plot(area_zscore, df_area_normalized)  
plt.xlabel(' Zscore Normalization')
plt.ylabel('Normalization technique used in data ')
plt.plot(a,a)
plt.show()
```

### scipy.stats.zscore(arr, axis=0, ddof=0) 
### function computes the relative Z-score of the input data, relative to the sample mean and standard deviation.

### Parameters :

arr : [array_like] Input array or object for which Z-score is to be calculated.

axis : Axis along which the mean is to be computed. By default axis = 0.

ddof : Degree of freedom correction for Standard Deviation.


Results : Z-score of the input data.
